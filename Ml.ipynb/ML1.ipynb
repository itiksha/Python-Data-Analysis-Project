```python
import sklearn
```

```python
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
```

```python
#Generate Sample Data
x, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)
```

```python
#Apply DBScan 
dbscan = DBSCAN(eps=0.8, min_samples=10)
labels = dbscan.fit_predict(x)
```

```python
#plot the results
plt.scatter(x[:,0], x[:,1], c=labels)
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('DBScan Clustering')
plt.show()
```

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
import numpy as np
import matplotlib.pyplot as plt
```

```python
num_samples_total = 1000
cluster_centers = [(3,3),(7,7)]
num_classes = len(cluster_centers)
eplison = 1.0
min_samples = 13

#Generate Data
X, y = make_blobs(n_samples= 1000, centers= cluster_centers, n_features= num_classes, center_box= (0,1), cluster_std=0.5)
plt.show()
```

```python
db = DBSCAN(eps= eplison, min_samples = min_samples).fit(X)
labels = db.labels_
```

```python
# Generate scatter plot for training data
colors = list(map(lambda x: '#3b4cc0' if x == 1 else '#b40426', labels))
plt.scatter(X[:,0], X[:,1], c=colors, marker="^", picker=True)
plt.title('Two clusters with data')
plt.xlabel('Axis X[0]')
plt.ylabel('Axis X[1]')
plt.show()
```
